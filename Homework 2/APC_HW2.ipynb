# Problem 1

```python
import numpy as np 
import matplotlib.pyplot as plt

xx = np.arange(-1, 1, 0.01) 
yy = np.arange(-1, 1, 0.01) 
x1, x2 = np.meshgrid(xx, yy) 

# Stable Node 
x1d_sn = -x1**3
x2d_sn = -x2**3
xd_sn = {'x1d':x1d_sn,'x2d':x2d_sn}

# Unstable Node
x1d_un = x1**3
x2d_un = x2**3
xd_un = {'x1d':x1d_un,'x2d':x2d_un}

# Stable Focus
x1d_sf = x1*(-x1**2 - x2**2) + x2*(x1**2 + x2**2)
x2d_sf = -x1*(x1**2 + x2**2) + x2*(-x1**2 - x2**2)
xd_sf = {'x1d':x1d_sf,'x2d':x2d_sf}

# Unstable Focus
x1d_uf = -x1*(-x1**2 - x2**2) - x2*(x1**2 + x2**2)
x2d_uf = +x1*(x1**2 + x2**2) - x2*(-x1**2 - x2**2)
xd_uf = {'x1d':x1d_uf,'x2d':x2d_uf}

# Center
x1d_c = -x2**3
x2d_c = x1**3
xd_c = {'x1d':x1d_c,'x2d':x2d_c}

# Saddle 
x1d_s = -x1**3
x2d_s = x2**3
xd_s = {'x1d':x1d_s,'x2d':x2d_s}

eq_pt_list = {'Stable Node':xd_sn, 
              'Unstable Node':xd_un,
              'Stable Focus':xd_sf,
              'Unstable Focus':xd_uf,
              'Center': xd_c,
              'Saddle': xd_s}


fig,([ax1, ax2, ax3], [ax4, ax5, ax6]) = plt.subplots(2, 3)
fig.set_figwidth(20)
fig.set_figheight(12)
fig.patch.set_facecolor('xkcd:white')
fig.suptitle('Problem 1')

ax_set = [ax1, ax2, ax3, ax4, ax5, ax6]

for i, key in enumerate(eq_pt_list):
    ax_set[i].title.set_text(key)
    data = eq_pt_list[key]
    ax_set[i].streamplot(x1, x2, data['x1d'], data['x2d'], density=[0.95, 1])
    ax_set[i].set_xlim([-1, 1])
    ax_set[i].set_ylim([-1, 1])
    ax_set[i].set_xlabel('x1')
    ax_set[i].set_ylabel('x2')

fig.savefig("HW2_P1.png")
```

# Problem 4

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# creating grid  
x = np.arange(-2, 2, 0.01) 
y = np.arange(-2, 2, 0.01) 
x1, x2 = np.meshgrid(x, y) 

# creating vector 
k = 1
x1d = x1*(k**2 - x1**2 - x2**2) + x2*(x1**2 + x2**2 + k**2)
x2d = -x1*(k**2 + x1**2 + x2**2) + x2*(k**2 - x1**2 - x2**2)

plt.figure(figsize=(8,8))
plt.streamplot(x1, x2, x1d, x2d, density=[1, 1])
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Problem 4, k = 1')
plt.savefig('HW2_P4.png')
```
